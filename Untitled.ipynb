```python
import matplotlib.pyplot as plt
import cv2
import glob
import os
import argparse
```

```python
#カスケードファイルの読み込み
cascade_path = "../opencv-master/data/haarcascades/haarcascade_eye.xml"
```

```python
# 基本的なモデルパラメータ
FLAGS = None

# 学習済モデルの種類
CASCADE = ["default","alt","alt2","tree","profile","nose"]

# 直接実行されている場合に通る(importされて実行時は通らない)
if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument(
        "--cascade",
        type=str,
        default="alt",
        choices=CASCADE,
        help="cascade file."
  )
    parser.add_argument(
        "--scale",
        type=float,
        default=1.3,
        help="scaleFactor value of detectMultiScale."
  )
    parser.add_argument(
        "--neighbors",
        type=int,
        default=2,
        help="minNeighbors value of detectMultiScale."
  )
    parser.add_argument(
        "--min",
        type=int,
        default=80,
        help="minSize value of detectMultiScale."
  )
    parser.add_argument(
        "--input_dir",
        type=str,
        default="./input/",
        help="The path of input directory."
  )
    parser.add_argument(
        "--move_dir",
        type=str,
        default="/done/",
        help="The path of moving detected files."
  )

# パラメータ取得と実行
FLAGS, unparsed = parser.parse_known_args() 
```

```python
#カスケードファイルを利用した検出器を定義
cascade = cv2.CascadeClassifier(cascade_path)
```

```python
def trimming_eye(data_path, save_path):
    i = 0
    
    for f in data_path:
        i += 1
        #画像ファイルを読み込み
        image_file = cv2.imread(f)
        #GrayScaleに変換
        img_gray = cv2.cvtColor(image_file, cv2.COLOR_BGR2GRAY)
        #detectMultiScaleで検出
        #for j in range(1,20):
        #    minValue = j * 5
        face_list = cascade.detectMultiScale(img_gray, minSize=(FLAGS.min, FLAGS.min))
        if len(face_list) == 2:
            break
        #face = cascade.detectMultiScale(img_gray, scaleFactor=FLAGS.scale, minNeighbors=FLAGS.neighbors, minSize=(FLAGS.min, FLAGS.min))
        #if len(face) > 0:
        #    break
            
        for (x,y,w,h) in face_list:
            img_cut = image_file[y:y+h, x:x+w]
        cv2.imwrite(save_path + '{0}.jpg'.format(i), img_cut)     
```

```python
files = glob.glob('archive/lfw-deepfunneled/not_smile/*.jpg')
file_path = 'archive/lfw-deepfunneled/not_smile_eye/'
trimming_eye(files, file_path)
```

```python
files = glob.glob('archive/lfw-deepfunneled/smile/*.jpg')
file_path = 'archive/lfw-deepfunneled/smile_eye/'
trimming_eye(files, file_path)
```
